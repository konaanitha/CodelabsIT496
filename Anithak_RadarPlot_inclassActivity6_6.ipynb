```python
import pandas as pd
import math

Aamatrix = pd.DataFrame(columns=['criteria','weight','ratingA','ratingB','scoreA','scoreB'])
print(Aamatrix)
```

```python
Aamatrix['criteria']=['risk','ROI','customerSatisfaction','feasibility','strategicAlignment','security','user']
print(Aamatrix)
```

```python
Aamatrix['weight']=[0.1,0.15,0.3,0.15,0.3,0.4,0.1]
print(Aamatrix)
```

```python
#Choose ratings

Aamatrix['ratingA'] = [4,1,2,1,5,4,3]
Aamatrix['ratingB'] = [3,4,2,3,2,5,4]

#calculate score
for index, row in Aamatrix.iterrows():

  Aamatrix['scoreA'][index] = row['ratingA'] * row['weight'] 
  Aamatrix['scoreB'][index] = row['ratingB'] * row['weight'] 

print(Aamatrix)

```

```python
#total scores

totalScoreA = 0
totalScoreB = 0

for index, row in Aamatrix.iterrows():
  totalScoreA += row['scoreA']
  totalScoreB += row['scoreB']

print('the total score of A is {:.2f} and for B is {:.2f}'.format(totalScoreA, totalScoreB))
```

```python
import numpy as np
import matplotlib.pyplot as plt

axisLocations = np.linspace(start=0, stop=2*np.pi,num=len(Aamatrix['ratingA']),endpoint=False)
axisLocations = np.concatenate((axisLocations,[axisLocations[0]]))
ratingsA = Aamatrix['ratingA']
ratingsB = Aamatrix['ratingB']

ratingsACircular = np.concatenate((ratingsA,[ratingsA[0]]))
ratingsBCircular = np.concatenate((ratingsB,[ratingsB[0]]))

plt.figure(figsize=(8,8))
plt.subplot(polar = True)

plt.plot(axisLocations,ratingsACircular,'o-', linewidth=2, label='In-house development')
plt.plot(axisLocations,ratingsBCircular,'d-', linewidth=2, label='Outsourc')

plt.title('Project Options Comparision', size = 20)
lines , labels = plt.thetagrids(np.degrees(axisLocations), labels= Aamatrix['criteria'])
plt.legend()
plt.show()


```
