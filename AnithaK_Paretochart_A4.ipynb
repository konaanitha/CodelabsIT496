```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
```

```python
df = pd.read_csv("/content/drive/MyDrive/IT496/ITSkills.csv")
print(df)
```

```python
df.index = df['Skill']
df = df.sort_values(by='frequency',ascending =False)
print(df)
```

```python
df["cumulativePercentage"] = df["frequency"].cumsum()/df["frequency"].sum() * 100

df["cumulativePercentage"] = df["cumulativePercentage"].apply(lambda x:round(x,2))
print(df)
```

```python
fig, ax = plt.subplots(figsize=(15,10))

ax.bar(df.index, df["frequency"], color="lightpink")

ax.set_xticklabels(df['Skill'], rotation=45)

ax2 = ax.twinx()

ax2.plot(df.index, df["cumulativePercentage"],color="C8", marker="D",ms=
         7, label = df["cumulativePercentage"])

ax2.axhline(80, color="blue", linestyle="dashed")
ax2.yaxis.set_major_formatter(PercentFormatter())

ax.tick_params(axis="y", colors="C5")
ax2.tick_params(axis="y", colors="C4")

for index, v in df.iterrows():
  label = round(v['cumulativePercentage'],1)
  plt.annotate(label, xy=(v['Skill'],v['cumulativePercentage']))
  
plt.show()
```
